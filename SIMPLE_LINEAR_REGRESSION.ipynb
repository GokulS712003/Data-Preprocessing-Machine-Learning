```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("/content/Salary_Data.csv")
```

```python
x=dataset.iloc[:,:-1].values
y=dataset.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 1/3, random_state = 0)
```

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(x_train,y_train)
```

```python
#so it is trained using the fit mtd, now let us predict test set results
y_pred=regressor.predict(x_test)#either ways we can do.. another coming in further steps
```

```python
#visulizing the training set results
#plotting the points
plt.scatter(x_train,y_train,color='red')
#plotting the line
plt.plot(x_train,regressor.predict(x_train),color='blue')

plt.title("salary vs experiece(training set)")
plt.xlabel('year of experience')
plt.ylabel('salary')
plt.show()
```

```python
#visulizing the test set results
#plotting the points
plt.scatter(x_test,y_test,color='red')
#plotting the line
plt.plot(x_train,regressor.predict(x_train),color='blue')

plt.title("salary vs experiece(test set)")
plt.xlabel('year of experience')
plt.ylabel('salary')
plt.show()
```
